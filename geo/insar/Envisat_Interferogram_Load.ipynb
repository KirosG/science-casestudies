This notebook demonstrates loading a locally stored ESA Environmental Satellite (Envisat) Interferogram

Download an interferogram in hdf5 format at http://www.unavco.org/data/imaging/sar/lts3/insar/INTERFEROGRAM/fielding/Envisat/77/Envisat_S2_077_0000_20100328-20100502_0035_-0200.h5

More information about Envisat can be found at https://earth.esa.int/web/guest/missions/esa-operational-eo-missions/envisat

Envisat Data provided by European Space Agency

Initial imports

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10,10)
import h5py
from skdaccess.utilities.file_browser import FileBrowser
```

Open Inteferogram file

For more information on the data format see https://winsar.unavco.org/portal/insar/

http://www.unavco.org/data/imaging/sar/lts3/insar/INTERFEROGRAM/fielding/Envisat/77/Envisat_S2_077_0000_20100328-20100502_0035_-0200.h5

```python
filename = FileBrowser()
filename.widget()
```

```python
f = h5py.File(filename.path,'r')
```

List keys in file

```python
list(f.keys())
```

List data in GEOCODE

```python
list(f['GEOCODE'])
```

Access data

```python
wrapped = f['GEOCODE']['wrapped_interferogram'][...]
```

Plot Interferogram

```python
plt.imshow(wrapped, cmap='jet');
plt.colorbar();
```
