This notebook demonstrates loading a locally stored JAXA Advanced Land Observing Satellite (ALOS) Interferogram.

Download an interferogram in hdf5 format at https://datapool.asf.alaska.edu/INTERFEROMETRY/A3/

More information about ALOS can be found at http://www.alos-restec.jp/en/staticpages/index.php/aboutalos

Dataset: Â© [JAXA/METI](http://www.eorc.jaxa.jp/ALOS/en/obs/palsar_strat.htm) ALOS PALSAR Interferogram 2010. Accessed through ASF DAAC 2017.

Initial imports

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10,10)
import h5py
from skdaccess.utilities.file_browser import FileBrowser
```

Open file

For more information on the data format see https://winsar.unavco.org/portal/insar/

https://datapool.asf.alaska.edu/INTERFEROMETRY/A3/ALPSR_23444_24786_0630.h5

```python
filename = FileBrowser()
filename.widget()
```

```python
f = h5py.File(filename.path,'r')
```

List root keys

```python
list(f.keys())
```

List subkeys

```python
list(f['ALPSR_23444_24786_0630'])
```

List data

```python
list(f['ALPSR_23444_24786_0630']['data'])
```

Access wrapped interferogram

```python
wrapped = f['ALPSR_23444_24786_0630']['data']['wrapped_interferogram'][...]
```

Plot inteferogram

```python
plt.imshow(wrapped,cmap='jet');
```
