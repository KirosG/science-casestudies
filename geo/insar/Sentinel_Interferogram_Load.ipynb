This notebook demonstrates loading a locally stored ESA Sentinel-1 Interferogram

The Interferogram may be downloaded at http://web-services.unavco.org/brokered/ssara/api/sar/search?platform=Sentinel-1&processingLevel=INTERFEROGRAM&output=map&start=2015-08-01&end=2015-09-01

More information about Sentinel-1 can be found at https://earth.esa.int/web/guest/missions/esa-operational-eo-missions/sentinel-1

Copernicus Sentinel data [2015]

Initial imports

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10,10)
import h5py
import numpy as np
from skdaccess.utilities.file_browser import FileBrowser
```

Open file downloaded from WInSAR

For more information on the data format see https://winsar.unavco.org/portal/insar/

http://www.unavco.org/data/imaging/sar/lts3/insar/INTERFEROGRAM/CunrenLiang/S1A/156/S1A_IW_156_5758_20150824-20150917_0024_00143.h5

```python
filename = FileBrowser()
filename.widget()
```

```python
f = h5py.File(filename.path,'r')
```

List keys in data

```python
list(f.keys())
```

List data in GEOCODE

```python
list(f['GEOCODE'])
```

Access Data

```python
wrapped = f['GEOCODE']['wrapped_interferogram'][...]
```

Plot interferogram

```python
plt.imshow(wrapped, cmap='jet');
plt.colorbar();
```
